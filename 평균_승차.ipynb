```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'

subway = pd.read_csv("종합운동장_행사_일주일전후_승차데이터.csv", encoding="cp949")
```

```python
subway
# 행사명, 행사날짜, 가까운 지하철역
```

```python
subway.info()
```

```python
for column in subway.columns[8:]:
    subway[column] = subway[column].astype(str)
    
# 승하차 인원 데이터 타입을 모두 문자열로 바꾼 후 float타입으로 바꿈. 예를 들어 1,100 -> 1100

for column in subway.columns[8:]:
    subway[column] = subway[column].str.replace(',', '').astype(float)

# 결과 출력
print(subway.dtypes)
```

```python
new_rows = []
for i in range(0, len(subway), 2):
    row = subway.iloc[i:i+2, 8:].astype(float).mean().tolist()
    new_rows.append(row)

# 새로운 행으로 이루어진 데이터프레임 생성
average_subway = pd.DataFrame(new_rows, columns=subway.columns[8:])

average_subway
```

```python
average_subway.to_csv('종합운동장_평균_지하철_승차_데이터.csv', index=False)
```
