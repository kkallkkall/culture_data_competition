```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'

bus_data = pd.read_excel("서울시버스노선별정류소정보(20230613).xlsx")

# 기존 버스 노선 파악

bus_station_jst = bus_data[bus_data["정류소명"].str.contains("잠실종합운동장")]

# 종합운동장을 지나면서 강남역도 지나는 버스 노선 탐색 

bus_routes = bus_station_jst["노선명"]
routes_with_gangnam = []

for route in bus_routes:
    route_df = bus_data[bus_data['노선명'] == route]
    has_gangnam = any(route_df['정류소명'].str.contains('강남역'))
    
    if has_gangnam:
        routes_with_gangnam.append(route)
        
    routes_with_gangnam_set = set(routes_with_gangnam)
    routes_with_gangnam = list(routes_with_gangnam_set)

print(f"'강남역'이라는 단어를 포함하는 정류소명을 지나는 노선들: {routes_with_gangnam}")
```

```python
!pip install folium

import folium 

m = folium.Map(location=[37.56816897719393, 126.89727772625356], zoom_start=12)

tiles = "CartoDB positron"
# 배경지도 타일 레이어를 지도에 추가하기
folium.TileLayer(tiles=tiles).add_to(m)
```

```python
colors = ['blue', 'red', 'green', 'orange','yellow','navy','purple']  # 각 노선에 대해 다른 색상을 지정

for i, route in enumerate(routes_with_gangnam):
    route_data = bus_data[bus_data['노선명'] == route]
    locations = []
    for index, row in route_data.iterrows():
        location = [row['Y좌표'], row['X좌표']]
        locations.append(location)
    color = colors[i % len(colors)]  # 색상 리스트에서 색상 선택
    folium.PolyLine(locations, color=color).add_to(m)

m
```
