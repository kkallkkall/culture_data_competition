```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'

bus_data = pd.read_excel("서울시버스노선별정류소정보(20230613).xlsx")

bus_station_WcStadium = bus_data[bus_data["정류소명"].str.contains("월드컵경기장")]
bus_station_WcStadium["노선명"].unique()
# 월드컵경기장을 지나면서 합정도 지나는 버스 노선 탐색 

bus_routes = bus_station_WcStadium["노선명"]
routes_with_hapjeong = []

for route in bus_routes:
    route_df = bus_data[bus_data['노선명'] == route]
    has_hapjeong = any(route_df['정류소명'].str.contains('합정'))
    
    if has_hapjeong:
        routes_with_hapjeong.append(route)

print(f"'합정'이라는 단어를 포함하는 정류소명을 지나는 노선들: {routes_with_hapjeong}")
```

```python
for route in routes_with_hapjeong:
    route_data = bus_data[bus_data['노선명'] == route]
    for index, row in route_data.iterrows():
        location = [row['Y좌표'], row['X좌표']]
        folium.Marker(location=location).add_to(m)
    folium.PolyLine(locations, color='blue').add_to(m)

m
```

```python
colors = ['blue', 'red', 'green', 'orange']  # 각 노선에 대해 다른 색상을 지정

for i, route in enumerate(routes_with_hapjeong):
    route_data = bus_data[bus_data['노선명'] == route]
    locations = []
    for index, row in route_data.iterrows():
        location = [row['Y좌표'], row['X좌표']]
        locations.append(location)
    color = colors[i % len(colors)]  # 색상 리스트에서 색상 선택
    folium.PolyLine(locations, color=color).add_to(m)

# 마커 제거
for child in m._children.values():
    if isinstance(child, folium.Marker):
        child.add_to(None)

m
```
