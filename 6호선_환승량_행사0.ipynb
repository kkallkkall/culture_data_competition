```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Malgun Gothic'

subway6 = pd.read_csv("6호선환승역.csv")
```

```python
grouped_data = subway6.groupby('환승역')
```

```python
group_data = grouped_data.get_group('태릉입구')
print(group_data)
```

```python
def sum_values(group):
    return group.iloc[:, 7:].sum()

# '환승역' 열 값을 기준으로 데이터프레임 그룹화
grouped_data = subway6.groupby('환승역')

# 그룹화된 데이터에 대해 sum_values 함수를 적용하여 값을 더함
group_sums = grouped_data.apply(sum_values)

# 결과 출력
group_sums
```

```python
group_sums.insert(0, '환승역', group_sums.index)
# 환승역명 추가
group_sums.reset_index(drop=True, inplace=True)
# 행이름 제거
group_sums
```

```python
group_sums.to_csv('station.csv', index=False)
```

```python
for i in range(1,21):
    plt.bar(group_sums['환승역'],group_sums.iloc[:,i])
    plt.xlabel('환승역')
    plt.ylabel('하차 인원') 
    plt.title(group_sums.iloc[:,i].name)
    plt.xticks(rotation = 80)

    plt.show()
    
# 행사 종료시간 21:50 이후 데이터인 22~23시 6호선 환승역 하차 인원 분석 결과 '합정','디지털미디어시티','석계','공덕' 에서 하차인원 많음
```
